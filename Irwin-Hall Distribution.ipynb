```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
num_uniform = [1,2,3,4,5]
sample_size = 10**6
num_bins = 60
```

```python
fig, axs = plt.subplots(2, 1, figsize = (10,8))
plt.subplots_adjust(top=0.8, hspace=0.4)
fig.suptitle("Распределение Ирвина-Холла \n Irwin-Hall Distribution", fontsize=14)

for n in num_uniform:
    
    x = [sum([np.random.uniform() for j in range(n)]) for k in range(sample_size)]
    bins = np.linspace(0,n,(n*num_bins)+1)
    
    axs[0].hist(x,bins = bins, density = True, histtype = 'step', label='n = '+str(n))
    axs[0].legend(loc=0)
    axs[0].set_title('Плотность распределения суммы n равномерных независимых случайных величин \n'
                     'PDF of the sum of n uniform independent random variables')

    axs[1].hist(x,bins = bins, density = True, cumulative = True, histtype = 'step', label='n = '+str(n))
    axs[1].legend(loc=0)
    axs[1].set_title('Функция распределения суммы n равномерных независимых случайных величин \n'
                     'CDF of the sum of n uniform independent random variables')
```

```python
num_uniform = [1,2,3,4,5]
sample_size = 10**6
num_bins = 60
    
fig, axs = plt.subplots(2, 1, figsize = (10,8))
plt.subplots_adjust(top = 0.8, hspace = 0.4)
fig.suptitle("Распределение Бейтса \n Bates Distribution", fontsize=14)

for n in num_uniform:
    
    x = [(1/n)*sum([np.random.uniform() for j in range(n)]) for k in range(sample_size)]
    bins = np.linspace(0,1,(n*num_bins)+1)
    
    axs[0].hist(x,bins = bins, density = True, histtype = 'step', label='n = '+str(n))
    axs[0].legend(loc=2)
    axs[0].set_title('Плотность распределения среднего значения n равномерных независимых случайных величин \n'
                     'PDF of the mean of n uniform independent random variables')

    axs[1].hist(x,bins = bins, density = True, cumulative = True, histtype = 'step', label='n = '+str(n))
    axs[1].legend(loc=2)
    axs[1].set_title('Функция распределения среднего значения n равномерных независимых случайных величин \n'
                     'CDF of the sum of n uniform independent random variables')
```
